```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 
plt.rcParams['font.sans-serif'] = ['SimHei']  # SimHei
plt.rcParams['axes.unicode_minus'] = False  # 

# 
g = 9.81  #  (m/s^2)
m = 0.2  #  (kg)
k = 0.1  #  (kg/s)

# 
t_max = 60  #  (s)
dt = 0.01  #  (s)
time = np.arange(0, t_max, dt)

# 
alpha = k / m
v = (m * g / k) * (1 - np.exp(-alpha * time))
s = (m * g / k) * (time - (m / k) * (1 - np.exp(-alpha * time)))

# 
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(time, v, label=' (m/s)')
plt.xlabel(' (s)')
plt.ylabel(' (m/s)')
plt.title('')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(time, s, label=' (m)', color='orange')
plt.xlabel(' (s)')
plt.ylabel(' (m)')
plt.title('')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
g = 9.81  #  (m/s^2)
m = 0.005  #  (kg)
k = 0.01  #  (kg/s)
rho_water = 1000  #  (kg/m^3)
V = 0.00001  #  (m^3)
v_flow = 1.0  #  (m/s)

# 
t_max = 60  #  (s)
dt = 0.01  #  (s)
time = np.arange(0, t_max, dt)

# 
effective_weight = (m - rho_water * V) * g
v_y = (effective_weight / k) * (1 - np.exp(-k * time / m))
s_y = (effective_weight / k) * (time - (m / k) * (1 - np.exp(-k * time / m)))

# 
s_x = v_flow * time

# 
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(time, v_y, label=' (m/s)')
plt.xlabel(' (s)')
plt.ylabel(' (m/s)')
plt.title('')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(time, s_y, label=' (m)', color='orange')
plt.xlabel(' (s)')
plt.ylabel(' (m)')
plt.title('')
plt.legend()

plt.tight_layout()
plt.show()

# 
plt.figure(figsize=(12, 6))
plt.plot(s_x, s_y, label='')
plt.xlabel(' (m)')
plt.ylabel(' (m)')
plt.title('')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# 
g = 9.81  #  (m/s^2)
rho_water = 1000  #  (kg/m^3)
v_flow = 1.0  #  (m/s)

# 
m_phone = 0.2  #  (kg)
V_phone = 0.00005  #  (m^3)
k_phone = 0.1  #  (kg/s)

# 
m_id = 0.005  #  (kg)
V_id = 0.00001  #  (m^3)
k_id = 0.01  #  (kg/s)

# 
t_max = 60  #  (s)

def compute_max_depth_and_drift(m, V, k, t_max, v_flow=None):
    effective_weight = (m - rho_water * V) * g
    y_max = (effective_weight / k) * t_max
    x_max = v_flow * t_max if v_flow is not None else 0
    return y_max, x_max

# 
y_max_phone, x_max_phone = compute_max_depth_and_drift(m_phone, V_phone, k_phone, t_max, v_flow)
print(f": {y_max_phone:.2f} m, : {x_max_phone:.2f} m")

# 
y_max_id, x_max_id = compute_max_depth_and_drift(m_id, V_id, k_id, t_max, v_flow)
print(f": {y_max_id:.2f} m, : {x_max_id:.2f} m")
```
