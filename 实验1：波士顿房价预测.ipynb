## 实验一：线性回归分析----波士顿房价预测实验

背景介绍

本项目将利用马萨诸塞州波士顿郊区的房屋信息数据训练和测试一个线性回归模型，并对模型的性能和预测能力进行测试。通过该数据训练后的好的模型可以被用来对房屋做特定预测---尤其是对房屋的价值。对于房地产经纪等人的日常工作来说，这样的预测模型被证明非常有价值。此项目的数据集来自UCI机器学习知识库(数据集已下线)。波士顿房屋这些数据于1978年开始统计，共506个数据点，涵盖了麻省波士顿不同郊区房屋14种特征的信息。

每条数据包含房屋以及房屋周围的详细信息。

* CRIM：城镇人均犯罪率；  　　　　　　　　　　　　
* ZN：住宅用地超过 25000 sq.ft. 的比例；
* INDUS：城镇非零售商用土地的比例；　　　　　　　
* CHAS：查理斯河空变量（如果边界是河流，则为1；否则为0）；
* NOX：一氧化氮浓度；　　　　　　　　　　　　　　
* RM：住宅平均房间数；
* AGE：1940 年之前建成的自用房屋比例；
* DIS：到波士顿五个中心区域的加权距离；
* RAD：辐射性公路的接近指数；　　　　　　　
* TAX：每 10000 美元的全值财产税率；
* PTRATIO：城镇师生比例；　　　　　　　　　　　　
* B：1000（Bk-0.63）^ 2，其中 Bk 指代城镇中黑人的比例。
* LSTAT：人口中地位低下者的比例。　　　　　　　
* MEDV：自住房的平均房价，以千美元计.

对于本项目，房屋的'RM'， 'LSTAT'，'PTRATIO'以及'MEDV'特征是必要的，其余特征可根据相关性进行选择。
'MEDV'特征的值已经过必要的数学转换，可以反映35年来市场的通货膨胀效应。

### 一、数据读取

#### 1、导入Python模块

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### 2、读入数据

```python
from sklearn.datasets import load_boston
boston = load_boston()
X,y = boston.data,boston.target
feature_names = boston.feature_names
print(X.shape)
print(y.shape)
```

查看数据

```python
df = pd.DataFrame(X, columns=boston.feature_names)
df.head()
```

```python
df.info()
```

### 二、模型训练与预测

划分数据集

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

模型训练

```python
import time
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

lr = LinearRegression()

start = time.perf_counter()#time.clock()
lr.fit(X_train, y_train)

y_train_pred = lr.predict(X_train)
train_error = mean_squared_error(y_train, y_train_pred)
print("训练误差(MSE)：",train_error)

y_pred = lr.predict(X_test)
test_error = mean_squared_error(y_test, y_pred)
print("测试误差(MSE)：",test_error)
print("R2 =",r2_score(y_test, y_pred))

# train_score = model.score(X_train, y_train)
# cv_score = model.score(X_test, y_test)
```

```python
#绘制直方图观察数据分布情况
plt.subplots(figsize=(30,10))
for i in range(len(df.columns)):
  plt.subplot(3,5,1+i)
  plt.hist(X[:,i],color='lightseagreen',width=2)
  plt.title(feature_names[i])
plt.show()
```
